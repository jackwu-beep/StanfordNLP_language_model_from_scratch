```python
import regex as re
html = "<h1>标题</h1><p>段落内容</p>"
# 匹配标签内的文本（非贪婪模式 .*? 避免过度匹配）
content = re.findall(r"<.*?>(.*?)</.*?>", html)
print(content)  # ['标题', '段落内容']
```

```python
testABC = 'testABC123'
content = re.findall(r""".*?(?=[A-Z])""", testABC)
content
```

```python
def isValidPhone(phone:str) -> bool:
    pattern = r"""^1[3-9]\p{N}{9}$"""
    return bool(re.match(pattern, phone))
isValidPhone('137一六440066')
```

```python
print(re.findall(r"\p{L}", "一六123"))
```

```python
PAT = r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?!\S)|\s+"""
re.findall(PAT, "some text that i'll pre-tokenize")
```

```python
it = re.finditer(PAT, "some text that i'll pre-tokenize")
[m.group() for m in it]
```
