```python
ans = 0
for i in range(1, 100):
    if i % 2:
        ans += i
    else:
        ans -= i
print(ans)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("数据2.3.csv")
# （1）查看数据集中的缺失值
missing = df.isnull()
print(df[missing.any(axis=1)])
```

```python
# （2）填充缺失值-用字符串'缺失数据'填充
df = pd.read_csv("数据2.3.csv")
df = df.fillna('缺失数据')
print(df)
```

```python
# （2）填充缺失值-用用变量均值或者中位数填充缺失数据
df = pd.read_csv("数据2.3.csv")
numeric_cols = df.select_dtypes(include=['number']).columns
for col in numeric_cols:
    # 这里用中位数，均数可改为 df[col].mean()
    df[col] = df[col].fillna(df[col].median())
print(df)
```
