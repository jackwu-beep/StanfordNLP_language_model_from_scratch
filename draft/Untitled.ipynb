```python
import regex as re
import os
from typing import BinaryIO
```

```python
file_path = '../data/TinyStoriesV2-GPT4-valid.txt'
```

```python
special_token = ['<|endoftext|>']
re_s_t = [re.escape(s) for s in special_token]
re_s_t[0]
```

```python
text = texts = ''
```

```python
with open(file_path, 'r') as f:
    text = f.read()
    texts = re.split("|".join(re_s_t), text)
    
```

```python
BPE = r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?:\S)|\s+"""
its = [re.finditer(BPE, t) for t in texts]
```

```python
ft = 'I have a pen.'
its_ft = re.finditer(BPE, ft)
print([it.group(0) for it in its_ft])  # ['I', ' have', ' a', ' pen', '.']
```

```python

test_it = its[0]
test_it_list = [i.group() for i in test_it]
```
